## Initial Data Processing

```r
raw_data_steepest = read.csv("./knapsack_problem_1/14_bit/knapsack_problem_1_14_bit_max_wt_90_sbc_op_steepest_node_counts_with_eigen_vals.csv")
raw_data_first_1 = read.csv("./knapsack_problem_1/14_bit/knapsack_problem_1_14_bit_max_wt_90_sbc_op_first_node_counts_with_eigen_vals_1.csv")
raw_data_first_2 = read.csv("./knapsack_problem_1/14_bit/knapsack_problem_1_14_bit_max_wt_90_sbc_op_first_node_counts_with_eigen_vals_2.csv")
raw_data_steepest_17_bit= read.csv("./knapsack_problem_1/17_bit/knapsack_problem_1_17_bit_max_wt_145_sbc_op_steepest_node_counts_with_eigen_vals.csv")
raw_data_first_17_bit= read.csv("./knapsack_problem_1/17_bit/knapsack_problem_1_17_bit_max_wt_145_sbc_op_first_node_counts_with_eigen_vals.csv")
```

```r
counts_steepest = raw_data_steepest %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent))
counts_first_1 = raw_data_first_1 %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent))
counts_first_2 = raw_data_first_2 %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent))
counts_steepest_17_bit = raw_data_steepest_17_bit %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent))
counts_first_17_bit = raw_data_first_17_bit %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent))
```
