```r
library(tidyverse)
```

```r
raw_data_knapsack_problem_FAN_first_14_bit_max_wt_77_1 = read.csv("./FAN_Validation/knapsack_problem_1/14_bit/max_wt_77/knapsack_problem_1_14_bit_max_wt_77_sbc_op_FAN_first_node_counts_with_eigen_vals_and_fitnesses_and_peaks_1.csv")
counts_knapsack_problem_FAN_first_14_bit_max_wt_77_1 = raw_data_knapsack_problem_FAN_first_14_bit_max_wt_77_1 %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent), fitness = mean(fitness), is_peak = first(is_peak))
```

## 14-bit knapsack max wt 77

```r
counts_knapsack_problem_FAN_first_14_bit_max_wt_77_1 %>%
  ggplot() +
  aes(x=fitness, y=eigen_cent) +
    geom_point()
```

```r
counts_knapsack_problem_FAN_first_14_bit_max_wt_77_1 %>%
  ggplot() +
  aes(x=fitness, y=count) +
    geom_point()
```

```r
counts_knapsack_problem_FAN_first_14_bit_max_wt_77_1 %>%
  filter(is_peak == "True") %>%
  ggplot() +
  aes(x=fitness, y=eigen_cent) +
    geom_point()
```

```r
counts_knapsack_problem_FAN_first_14_bit_max_wt_77_1 %>%
  filter(is_peak == "True") %>%
  ggplot() +
  aes(x=fitness, y=count) +
    geom_point()
```
