```r
library(tidyverse)
```

```r
raw_data_SAN_steepest_14_bit = read.csv("./SAN_Validation/knapsack_problem_1/14_bit/max_wt_90/knapsack_problem_1_14_bit_max_wt_90_sbc_op_steepest_node_counts_with_eigen_vals_and_fitnesses_and_peaks.csv")
counts_SAN_steepest_14_bit = raw_data_SAN_steepest_14_bit %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent), fitness = mean(fitness), is_peak = first(is_peak))
```

```r
raw_data_FAN_first_1_14_bit = read.csv("./FAN_Validation/knapsack_problem_1/14_bit/max_wt_90/knapsack_problem_1_14_bit_max_wt_90_sbc_op_FAN_first_node_counts_with_eigen_vals_and_fitnesses_and_peaks_1.csv")
counts_FAN_first_1_14_bit = raw_data_FAN_first_1_14_bit %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent), fitness = mean(fitness), is_peak = first(is_peak))
```

```r
raw_data_RAN_random_14_bit = read.csv("./RAN_Validation/knapsack_problem_1/14_bit/max_wt_90/knapsack_problem_1_14_bit_max_wt_90_sbc_op_random_node_counts_1_with_RAN_eigen_vals_and_fitnesses_and_peaks_1.csv")
counts_RAN_random_14_bit = raw_data_RAN_random_14_bit %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent), fitness = mean(fitness), is_peak = first(is_peak))
```

```r
counts_SAN_steepest_14_bit %>%
  filter(is_peak == "True") %>%
  ggplot() +
  aes(x=fitness, y=eigen_cent) +
    geom_point()
```

```r
counts_SAN_steepest_14_bit %>%
  filter(is_peak == "True") %>%
  ggplot() +
  aes(x=fitness, y=count) +
    geom_point()
```
