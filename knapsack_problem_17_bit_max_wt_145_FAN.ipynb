```r
library(tidyverse)
```

```r
raw_data_knapsack_problem_FAN_first_17_bit_max_wt_145_1 = read.csv("./FAN_Validation/knapsack_problem_1/17_bit/knapsack_problem_1_17_bit_max_wt_145_sbc_op_FAN_first_node_counts_with_eigen_vals_and_fitnesses_and_peaks_1.csv")
counts_knapsack_problem_FAN_first_17_bit_max_wt_145_1 = raw_data_knapsack_problem_FAN_first_17_bit_max_wt_145_1 %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent), fitness = mean(fitness), is_peak = first(is_peak))
```

## 17 bit max wt 145

```r
counts_knapsack_problem_FAN_first_17_bit_max_wt_145_1 %>%
  ggplot() +
  aes(x=fitness, y=eigen_cent) +
    geom_point()
```

```r
counts_knapsack_problem_FAN_first_17_bit_max_wt_145_1 %>%
  ggplot() +
  aes(x=fitness, y=count) +
    geom_point()
```

```r
counts_knapsack_problem_FAN_first_17_bit_max_wt_145_1 %>%
  filter(is_peak == "True") %>%
  ggplot() +
  aes(x=fitness, y=eigen_cent) +
    geom_point()
```

```r
counts_knapsack_problem_FAN_first_17_bit_max_wt_145_1 %>%
  filter(is_peak == "True") %>%
  ggplot() +
  aes(x=fitness, y=count) +
    geom_point()
```
