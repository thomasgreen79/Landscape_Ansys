# Analyzing NK_problem FL structure

```r
raw_data_SAN_steepest_14_bit = read.csv("./SAN_Validation/nk_problem/14_bit/kval_03/seed_19/nk_problem_14_bit_kval_03_seed_19_sbc_op_steepest_node_counts_with_SAN_eigen_vals_and_fitnesses_and_peaks.csv")
counts_SAN_steepest_14_bit = raw_data_SAN_steepest_14_bit %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent), fitness = mean(fitness), is_peak = first(is_peak))

raw_data_FAN_first_14_bit = read.csv("./FAN_Validation/nk_problem/14_bit/kval_03/seed_19/nk_problem_14_bit_kval_03_seed_19_sbc_op_first_node_counts_with_FAN_eigen_vals_and_fitnesses_and_peaks_1.csv")
counts_FAN_first_14_bit = raw_data_FAN_first_14_bit %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent), fitness = mean(fitness), is_peak = first(is_peak))
```

```r
ggplot(counts_SAN_steepest_14_bit) +
  aes(x=eigen_cent, y=count) +
  geom_point()
```

```r
counts_SAN_steepest_14_bit %>%  
#   filter(is_peak == "True") %>%
  ggplot() +
  aes(x=fitness, y=count) +
    geom_point()
```
