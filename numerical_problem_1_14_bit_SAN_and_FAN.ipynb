```r
library(tidyverse)
```

```r
raw_data_numerical_1_steepest_14_bit = read.csv("./SAN_Validation/numeric_problem_1/14_bit/numerical_problem_1_14_bit_sbc_op_steepest_node_counts_with_eigen_vals.csv")
counts_numerical_1_steepest_14_bit = raw_data_numerical_1_steepest_14_bit %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent))
```

```r
raw_data_numerical_1_first_14_bit_1 = read.csv("./SAN_Validation/numeric_problem_1/14_bit/numerical_problem_1_14_bit_sbc_op_first_node_counts_with_eigen_vals_1.csv")
counts_numerical_1_first_14_bit_1 = raw_data_numerical_1_first_14_bit_1 %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent))
```

```r
raw_data_numerical_1_first_14_bit_2 = read.csv("./SAN_Validation/numeric_problem_1/14_bit/numerical_problem_1_14_bit_sbc_op_first_node_counts_with_eigen_vals_2.csv")
counts_numerical_1_first_14_bit_2 = raw_data_numerical_1_first_14_bit_2 %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent))
```

```r
raw_data_numerical_1_FAN_first_14_bit_1 = read.csv("./FAN_Validation/numeric_problem_1/14_bit/numerical_problem_1_14_bit_sbc_op_FAN_first_node_counts_with_eigen_vals_and_fitnesses_and_peaks_1.csv")
counts_numerical_1_FAN_first_14_bit_1 = raw_data_numerical_1_FAN_first_14_bit_1 %>% 
  group_by(id) %>%
  summarise(count = sum(count), eigen_cent = mean(eigen_cent), fitness = mean(fitness), is_peak = first(is_peak))
```
